## Split Data

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('HappinessScore',axis=1)
y=df['HappinessScore']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## standardization

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train= sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

## Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
M1=LinearRegression()
M1.fit(X_train,y_train)
P1=M1.predict(X_test)
print(mean_absolute_error(y_test,P1))
print('\n')
print(mean_squared_error(y_test,P1))
print('\n')
print(np.sqrt(mean_squared_error(y_test,P1)))
print('\n')
print(r2_score(y_test,P1))
```

## Linear Regression dig 2

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
M2=make_pipeline(PolynomialFeatures(degree=2),LinearRegression())
M2.fit(X_train,y_train)
P2=M2.predict(X_test)
print(mean_absolute_error(y_test,P2))
print('\n')
print(mean_squared_error(y_test,P2))
print('\n')
print(np.sqrt(mean_squared_error(y_test,P2)))
print('\n')
print(r2_score(y_test,P2))
```

## Lasso

```python
from sklearn.linear_model import Lasso
```

```python
M3=make_pipeline(PolynomialFeatures(degree=3),Lasso())
M3.fit(X_train,y_train)
P3=M3.predict(X_test)
print(mean_absolute_error(y_test,P3))
print('\n')
print(mean_squared_error(y_test,P3))
print('\n')
print(np.sqrt(mean_squared_error(y_test,P3)))
print('\n')
print(r2_score(y_test,P3))
```

## Random forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
M4=RandomForestRegressor()
M4.fit(X_train,y_train)
P4=M4.predict(X_test)
print(mean_absolute_error(y_test,P4))
print('\n')
print(mean_squared_error(y_test,P4))
print('\n')
print(np.sqrt(mean_squared_error(y_test,P4)))
print('\n')
print(r2_score(y_test,P4))
```

## Neural Network

```python
from sklearn.neural_network import MLPRegressor
```

```python
M5=MLPRegressor(hidden_layer_sizes=(500,))
M5.fit(X_train,y_train)
P5=M5.predict(X_test)
print(mean_absolute_error(y_test,P5))
print('\n')
print(mean_squared_error(y_test,P5))
print('\n')
print(np.sqrt(mean_squared_error(y_test,P5)))
print('\n')
print(r2_score(y_test,P5))
```
